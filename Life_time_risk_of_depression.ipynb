```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Share - Question: mh7a '], bins=30, kde=True)
plt.title('Distribution of Anxiety or Depression Share')
plt.xlabel('Share (%)')
plt.ylabel('Frequency')
plt.savefig("Distribution of Anxiety or Depression Share.png")
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=df, x='Year', y='Share - Question: mh7a ', hue='country', legend=None)
plt.title('Trend of Anxiety or Depression Share Over the Years')
plt.xlabel('Year')
plt.ylabel('Share (%)')
plt.savefig("Trend of Anxiety or Depression Share Over the Years.png")
plt.show()
```

```python
# Select numeric columns
numeric_df = df[['Year', 'Share - Question: mh7a ']]

# Compute correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.savefig("Correlation Heatmap.png")
plt.show()
```
